```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# reads article text from file
with open("article.txt", "r", encoding="utf-8") as file:
    text = file.read()
```

```python
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="white",
    stopwords={"and", "to", "at", "a", "the", "of", "in", "is"}
).generate(text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.savefig("wordcloud.png")
plt.show()
```

In this task, I selected an article from news.err.ee and saved its content into a text file.
Using Python, I read the article text from the file and generated a word cloud with the WordCloud library.
Common stopwords were removed to highlight the most frequent and meaningful words.
The word cloud visually represents the key topics discussed in the article.
